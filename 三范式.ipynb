```python
/***************************数据库模块三范式****************************************
第一范式:把属性拆分成不可再生的最小单元，如联系方式可拆分成电话和邮箱,相对最小,姓名可以不分
    缺点：加大查询难度

第二范式:实体唯一性，不存在一个实体出现在多行的状况
    缺点：1.数据冗余 2.维护管理困难，牵一发动全身
    解决方案：分表处理，需要的时候再按要求合表，跟多文件模块编程一个道理，因为表之间没有关系，所以要拆分成两张独立的表、
            再找他们的关系，建立第三张关系表

第三范式:不存在中间依赖，列与列之间中间关系曲折
    缺点：同二
    解决方案：分表处理，这里表与表之间本身就有关系，直接拆封分成两个表就可以,表与表之间有关系列
**********************************************************************************/

```

不符合第三范式，存在中间依赖，即专业编号也依赖于学号，而专业名称依赖于专业编号，但专业名称和学号在一个表


```python
/************************表与表之间的关系*************************/
-- 一对多（专业表、学生表）
/*一个专业对应多个学生，那么就在学生表里面添加给外键，专业编号，关联专业表*/
/*
一对一，可以合成一张表或者公用一个主键
一对多，可以多的那个引用一的主键作为外键
多对多，可以新建一张表存储两者的关系
*/
create table Profession
(
    Pro_id int primary key identity(1,1) , --专业编号
    Pro_name varchar(50) not null
)

create tbale student
(
    Stu_id int primary key identity(1,1) --学生编号
    Pro_id int primary key identity(1,1) --这里用来关联专业表，自动编号，identity（1,1）表示起点和步长
    Stu_name varchar(50) not null,
    Stu_sex varchar(2) not null --学生性别 
)

/*这里必须先有Profession后才有Student表*/

--测试
insert into Profession(ProName) value('语文')
insert into Profession(ProName) value('计算机')
insert into Student(Pro_id,StuName,Stu_sex) value (1,'Jack','M')
insert into Student(Pro_id,StuName,Stu_sex) value (2,'Sary','W')


--一对一
/*一对多也可以表示一对一，在插入时候值插入一条，也有自己的处理方式*/
/*这里采用填入编号的方式，而不是自动编号，避免因为填入顺序不同导致的信息错乱*/
create table StudentBasicInfo --学生基本信息
(
    StuNo varchar(20) primary key not null,  --学号
    StuName varchar(20) not null,
    StuSex varchar (1) not null
)

create table StudentDetailInfo --学生详细信息
(
    StuNo varchar(20) primary key not null,
    StuQQ varchar(20),
    StuPhone varchar(20),
    StuMail varchar(100),
    StuBirth date 
)


--多对多(学生信息，课程信息)
/------------*再添加一张表*---------------------/
create table StudentBasicInfo --学生基本信息
(
    StuNo varchar(20) primary key not null,  --学号
    StuName varchar(20) not null,
    StuSex varchar (1) not null
)
create table Course 
(
    CourseId int primary key identity(1,1),
    CourseName varchar(30) not null,
    courseContent text --课程介绍
)
create Tabel Exam
(
    ExamId int primary key identity(1,1),
    StuNo  int not null ,
    CourseId int not null,
    Scorse int not null 
)











```
